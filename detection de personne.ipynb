```python
from ultralytics import YOLO
import cv2

# 🔹 Charger ton modèle personnalisé
model=YOLO(r"C:\Users\asmae\Downloads\best (1).pt")
# 🔹 Ouvrir la webcam (0 = caméra par défaut)
cap = cv2.VideoCapture(1)

if not cap.isOpened():
    print("❌ Erreur : impossible d’accéder à la webcam.")
    exit()

print("✅ Webcam connectée. Appuie sur 'q' pour quitter.")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # 🔹 Lancer la détection avec ton modèle
    results = model(frame, verbose=False)

    # 🔹 Dessiner les boîtes de détection
    for result in results:
        for box, cls, conf in zip(result.boxes.xyxy, result.boxes.cls, result.boxes.conf):
            if conf > 0.3:  # Confiance minimale de 50%
                x1, y1, x2, y2 = map(int, box)
                label = f"Person {conf:.2f}"
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(frame, label, (x1, y1 - 10),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

    # 🔹 Afficher la vidéo en temps réel
    cv2.imshow("Détection de camions - Modèle personnalisé", frame)

    # 🔹 Quitter avec la touche 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 🔹 Fermer la webcam et les fenêtres
cap.release()
cv2.destroyAllWindows()
```
